### Data Cleaning Transformation

```python
# drop missing values
yanki_df.dropna(subset=['Order_ID', 'Customer_ID'], inplace=True)
```

```python
yanki_df.info()
```

```python
# Conver Order_date from string to datetime object
yanki_df['Order_Date'] = pd.to_datetime(yanki_df['Order_Date'])
```

```python
yanki_df.info()
```

```python
#customer table
customer_df = yanki_df[['Customer_ID', 'Customer_Name','Email', 'Phone_Number']].copy().drop_duplicates().reset_index(drop=True)
customer_df.head()
```

```python
# product table
product_df = yanki_df[['Product_ID',
       'Product_Name', 'Brand','Price']].copy().drop_duplicates().reset_index(drop=True)
product_df.head()
```

```python
# shipping_address table
shipping_address_df = yanki_df[['Customer_ID','Shipping_Address', 'City', 'State', 'Country',
       'Postal_Code',]].copy().drop_duplicates().reset_index(drop=True)
# To get the shipping id, I will rename the index to column called ahipping ID

shipping_address_df.index.name = 'Shipping_ID'
shipping_address_df = shipping_address_df.reset_index()
shipping_address_df.head()
```

```python
# order table
orders_df =yanki_df[['Order_ID', 'Product_ID','Customer_ID', 'Quantity', 'Total_Price',
       'Order_Date']].copy().drop_duplicates().reset_index(drop=True)

orders_df.head()
```

```python
# payment method table
payment_method_df = yanki_df[['Order_ID', 'Payment_Method',
       'Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)

payment_method_df.head()
```
